##  Parâmetros nomeados





Construção de um dataFrame passando parâmetros nomeados **data, columns e index**


`data:` O parâmetro data é usado para fornecer os dados que preencherão o DataFrame. Os dados podem ser fornecidos em diferentes formatos

`columns:` O parâmetro columns é usado para especificar os nomes das colunas do DataFrame. Ele é opcional e, se não for fornecido, o Pandas atribuirá automaticamente nomes de colunas numéricas.

`index:` O parâmetro index é usado para especificar os rótulos das linhas do DataFrame. Ele também é opcional. Se não for fornecido, o Pandas atribuirá automaticamente rótulos de linha numéricos.







SINTAXE

`novoDataFrame = pd.DataFrame(data=[],columns=[],index=[])`


```python
## criar um dataFrame com 10 linhas
dfs = pd.DataFrame(
    data=[],
    columns=['Lógica de programação', 'Python', 'DS', 'HTML e CSS', 'JavaScript', 'React'],
    index=[list(range(1,11))]
)
dfs
```

# Manipulação de colunas

A biblioteca __Pandas__ possui muitas funções e métodos implementados que nos ajudam, e muito, em nosso dia-a-dia na análise exploratória de dados.

Nesta aula, vamos aprender a utilizar os mais comuns conforme estão listados abaixo:

* `rename` : renomeia os 'rótulos' dos eixos (índices ou colunas)

* `apply` : aplica uma determinada função ao longo de um eixo quando for chamado diretamente pelo _DataFrame_ ou aplica uma função numa _Series_ em questão

* `map` : "mapeia" uma _Series_ aplicando determinada função a cada elemento desta

* `agg` : agrega uma ou mais operações à uma _Series_

* `min` : retorna o valor mínimo de uma _Series_

* `max` : retorna o valor máximo de uma _Series_

* `sum` : retorna a soma dos valores de uma _Series_

* `mean` : retorna a média dos valores de uma _Series_

* `count` : conta as células não vazias de cada linha ou coluna

* `value_counts` : retorna uma _Series_ contendo a contagem de linhas exclusivas no _DataFrame_

* `sort_values` : classifica / ordena os valores ao longo de qualquer eixo

* `insert` : insire uma nova coluna no _DataFrame_ no local especificado

* `drop` : remova linhas ou colunas especificando diretamente seus nomes

## Inserindo

### Inserir uma nova coluna



SINTAX

`nomeDataFrame.insert(loc, column, value, allow_duplicates=False)`



*   loc - valor para posição da coluna ( deve ser inteiro)

*   column - nome da coluna

* value - esse valor pode ser uma Series, um array ou um resultado de algum cálculo

* allow_duplicates: por padrão é False, mas pode ser modificado.




```python
#adicionando uma coluna nova em um indice específico
#DataFrame.insert()
alunos = ['Maria','Paulo','João','Ana', 'Lívia', 'Fernanda','Telmo','Otávio', 'Lúcia', 'Maria']
dfs.insert(0, 'Nome Aluno', alunos , True)
dfs

```

### Nova coluna pela média

```python
#adicionando uma coluna nova
dfs['Media']=dfs.mean(axis='columns')
dfs
```

## Removendo



```python
#removendo a última coluna
dfs.drop(['React'], axis='columns',inplace= True)
```

## Populando o dataFrame

Importaremos a biblioteca **numpy** para nos auxiliar a popular o dataFrame com notas dentro de um intervalo definido, ao longo das colunas.
Usaremos o método `apply` e uma função lambda para nos auxiliar nessa tarefa.

sintaxe

`dataFrame.apply( função de Callback, axis= valor )`

`axis` pode receber `"index"` ou `0` para as linhas e `"columns"` ou `1 `para as colunas


```python
import numpy as np
dfs = dfs.apply(lambda nota: np.random.uniform(low=5.0, high=10.0, size=10), axis=0)
dfs
```

### Nova coluna com a média

```python
dfs['Média']= dfs.mean(axis='columns')
dfs
```

### Nova columa a partir de um teste lógico

A partir da coluna média, nós vamos criar uma nova coluna para avaliar a situação de aprovação ou reprovação
Para esse procedimento, vamos levar em conta apenas uma coluna e não todas.
o método mais adequado para esse procedimento é o * `map`.

"mapeia" uma _Series_ aplicando determinada função a cada elemento desta

`df[Nome coluna].map(função)`

```python
dfs['Situação'] = dfs['Média'].map(lambda media: 'AP' if media>7 else ("RC" if media>4 else 'RP'))
dfs
```

Ainda é possíevl criar uma nova coluna a partir de alguma operação aritmética

```python
dfs['teste'] = dfs['Python']*dfs['DS']
dfs
```

```python
dfs.drop(['teste'], axis='columns',inplace= True)
dfs
```

## Renomeando colunas

* converter para maiúsculas : `str.upper`
* converter para minúsculas: `str.lower`
* inverter maiúsculas e minúsculas: ` str.swapcase`

# Ordenação



SINTAXE

`nomeDataFrame.sort_values(by= nomeColuna ou [nome1, nome2], ascending=True  )`

Ordenação por padrão é realizada de forma ascendente
**Parâmetros:**

`by`
* único ou  lista

`axis`
*    0 ou 'indice' para linhas
*    1 ou 'columns' para colunas

`ascending`
* valor boleano que classifica em ordem crescente se True

`inplace`
* valor booleano. Faz as alterações no próprio dataframe transmitido, se True.





```python
# ordenação da menor versão para a maior
linguagens_df.sort_values(by='Versão')
```

```python
# ordenação da linguagem com mais profissionais para a menor (decrescente)
linguagens_df.sort_values(by='desenvolvedores', ascending=False)
```

## Insira uma coluna antes da coluna de valor total que mostre o texto Ticket alto, médio ou baixo, tendo como parâmetro :
* baixo <= 500.00
* medio <= 1500.00
