### 12- Na coluna nota os clientes avaliam a experiência com o restaurante pontuando de 0 a 5.
Com base na tabela abaixo crie uma coluna para os novos parâmetrosimage-asset.png

```python
def avaliacao(pontuacao):
    if pontuacao == 0:
        return 'muito insatisfeito'
    elif pontuacao == 1 or pontuacao ==2:
        return 'pouco insatisfeito'
    elif pontuacao == 3:
        return 'neutro'
    elif pontuacao == 4:
        return 'pouco satisfeito'
    elif pontuacao == 5:
        return 'muito satisfeito'
    else:
        return 'não classificado'
```

```python
condicao = lambda nota: 'muito insatisfeito' if (nota<1.0) else (#\
        'pouco insatisfeito' if (nota<=2.0) else ('neutro' if (nota<4.0)
         else ('pouco insatisfeito' if (nota<5.0) else 'muito satisfeito')))
```

```python
df
```

```python
df.columns
```

```python
df['avaliacao'] = df[' nota'].map(condicao)
df
```
